#### (c) (3 points)

**Use Python to fit a simple linear regression model relating the budget of internet video (x) to the amount of sales (y). Paste the Python output for the least squares estimate of the slope coefficient, and the least squares estimate of the intercept.**

```python
n = df.shape[0]
x_bar = df.mean()[0]; y_bar = df.mean()[1]
x_square = df['Advertisement']**2
xy = df['Advertisement'] * df['sales']
s_xx = x_square.sum() - n * (x_bar ** 2)
s_xy = xy.sum() - n * x_bar * y_bar
beta1 = s_xy / s_xx
beta0 = y_bar - beta1 * x_bar



plt.xlabel('Budget of Advertisement')
plt.ylabel('Sales')
plt.plot(df['Advertisement'], beta0 + beta1*df['Advertisement'],color = 'red')
plt.show()

print("y = ", beta0, " + ", beta1, "x")
```
